```python
# 변수 x의 값이 10보다 크거나 같으면 Pass 출력
x=11;
if x>=10:
    print('pass')
```

```python
today_temp=10;
if today_temp>=23:
    print('아이스 아메리카노')
else :
    print('따뜻한 아메리카노')
```

```python
weather="맑음"
temp_today=20;

if weather=="맑음":
    if temp_today >=20:
        print('아이스')
    else :
        print('따뜻한')
else:
    print('카푸치노')
    
```
